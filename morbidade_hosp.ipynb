```python
import plotly.graph_objects as go
import numpy as np

df = pd.read_csv('./dataset.csv', delimiter=';', thousands='.', decimal=',')

df = df[:-1]

anos = df['Ano atendimento']
internacoes = df['Internacoes']
valor_total = df['Valor_total']

x_mesh, y_mesh = np.meshgrid(anos, internacoes)

# Converter os valores do eixo Z para um array 2D
z_array = np.array([anos, valor_total])

fig = go.Figure(data=[
    go.Surface(
        x=x_mesh,
        y=y_mesh,
        z=z_array,
        #colorscale='Viridis',
        #showscale=True
    )
])

# Configurar o layout do gráfico
fig.update_layout(
    title='Gráfico 3D',
    scene=dict(
        xaxis_title='Ano',
        yaxis_title='Internações',
        zaxis_title='Valor Total'
    )
)

# Exibir o gráfico
fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np

df = pd.read_csv('./dataset.csv', delimiter=';', thousands='.', decimal=',')

df = df[:-1]

anos = df['Ano atendimento']
internacoes = df['Internacoes']
valor_total = df['Valor_total']
obitos = df['Obitos']

fig = go.Figure(data=[go.Mesh3d(x=anos,
                                y=obitos,
                                z=internacoes,
                                colorscale='Viridis',
                                showscale=True
                                )])

fig.update_layout(
    title='Gráfico 3D',
    scene=dict(
        xaxis_title='Ano',
        yaxis_title='Óbitos',
        zaxis_title='Internações'
    )
)

fig.show()
```
